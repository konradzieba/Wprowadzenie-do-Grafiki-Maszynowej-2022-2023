```python
from PIL import Image
import numpy as np
from PIL import ImageChops, ImageOps, ImageShow, ImageFilter
from PIL import ImageStat as stat
import matplotlib.pyplot as plt
```

```python
font = {
        'family': 'serif',
        'color':  'darkred',
        'weight': 'bold',
        'size': 28,
        }
```

```python
#Zadanie 1
obraz1 = Image.open('obraz1.png')
r, g, b, a = obraz1.split()
print('Obraz (przed konwersją) tryb:', obraz1.mode)
obraz2 = obraz1.convert('RGB')
print('Obraz2 (po konwersji) tryb:', obraz2.mode)
```

```python
obraz3 = Image.new('RGB', obraz1.size, (255,255,255))
obraz3.paste(obraz2, (0,0), a)
print('Tryb utworzonego obrazu (obraz3):', obraz3.mode)
obraz3.save('obraz3.png')
obraz3.show()
```

```python
img_diff = ImageChops.difference(obraz2, obraz3)
img_diff.show()
```

```python
img_cmyk = obraz3.convert('CMYK')
img_ycbcr = obraz3.convert('YCbCr')
img_hsv = obraz3.convert('HSV')
```

```python
plt.figure(figsize=(35, 20))
plt.subplot(1, 3, 1)  # ile obrazów w pionie, ile w poziomie, numer obrazu
plt.imshow(img_cmyk)
plt.axis('off')
plt.title('Obraz w trybie CMYK', fontdict=font)
plt.subplot(1, 3, 2)  # ile obrazów w pionie, ile w poziomie, numer obrazu
plt.imshow(img_ycbcr)
plt.axis('off')
plt.title('Obraz w trybie YcbCr', fontdict=font)
plt.subplot(1, 3, 3)  # ile obrazów w pionie, ile w poziomie, numer obrazu
plt.imshow(img_hsv)
plt.axis('off')
plt.title('Obraz w trybie HSV', fontdict=font)
plt.savefig('fig1.png')
plt.show()
```

```python
img_cmyk_c, img_cmyk_m, img_cmyk_y, img_cmyk_k = img_cmyk.split()
img_ycbcr_1, img_ycbcr_2, img_ycbcr_3 = img_ycbcr.split() 
img_hsv_h, img_hsv_s, img_hsv_v = img_hsv.split()
```
